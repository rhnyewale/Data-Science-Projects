# Arima Model

```python
emissionArima = emissionMain.copy()
dropColsArima = ['Coal, Including Coal Coke Net Imports, CO2 Emissions','Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions','Distillate Fuel Oil, Excluding Biodiesel, CO2 Emissions','Motor Gasoline, Excluding Ethanol, CO2 Emissions','Residual Fuel Oil CO2 Emissions','Petroleum, Excluding Biofuels, CO2 Emissions','Other Carbon Emission']
emissionArima = emissionArima.drop(dropColsArima, axis=1)
emissionArima = emissionArima.set_index('Month')
emissionArima
```

```python
trainArima = emissionArima
testArima = emissionArima[480:564]
plt.plot(trainArima)
plt.plot(testArima)
```

```python
testArima
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams
%matplotlib inline

trainLog = np.log(trainArima)
plt.figure(figsize=(20,7))
plt.plot(trainLog)
```

```python
rolling_mean_trainArima = trainLog.rolling(window=12).mean()
rolling_mean_trainArima
```

```python
original = plt.plot(trainLog, color='blue', label='Original')
mean = plt.plot(rolling_mean_trainArima, color='red', label='Rolling mean')

plt.legend(loc='best')
plt.title("Rolling Mean")
plt.show(block=False)
```

```python
# null-hypothesis : Data is Stationary

from statsmodels.tsa.stattools import adfuller

print('Results of Dickey-Fuller Test')
dftest = adfuller(trainLog['Total Energy CO2 Emissions'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#lags used', 'no. of observations used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value(%s)' %key] = value
    
print(dfoutput) 
```

```python
trainArimaShift = trainLog - trainLog.shift()
plt.plot(trainArimaShift)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    #determining rolling ststistics
    movingAverage=timeseries.rolling(window=10).mean()
    movingStd= timeseries.rolling(window=1).std()
    
    #Plot rolling statistics
    orig = plt.plot(timeseries, color= 'blue', label = 'Original')
    mean = plt.plot(movingAverage, color = 'black', label = 'Rolling mean')
    plt.legend(loc='best')
    plt.title('Rolling mean')
    plt.show(block=False)
    
    
    #Dickey Fuller Test
    print('Results of Dickey Fuller Test')
    dftest = adfuller(timeseries["Total Energy CO2 Emissions"], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#lags used', 'no. of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value(%s)' %key] = value
    print(dfoutput)
```

```python
trainArimaShift.dropna(inplace=True)
test_stationarity(trainArimaShift)
```

```python
#ACF and PACF plots
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(trainArimaShift, nlags=40)
lag_pacf = pacf(trainArimaShift, nlags=40, method='ols')
#ols is Ordinary Least Squared method
#Plot ACF:
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.stem(lag_acf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.stem(lag_pacf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(trainArimaShift)), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(trainLog, order=(2,1,2))
#(p,d,q)
results_ARIMA = model.fit()
plt.plot(trainArimaShift)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f' % sum((results_ARIMA.fittedvalues-trainArimaShift['Total Energy CO2 Emissions'])**2))
print('Plotting AR Model')
```

# SARIMA Model

```python
emissionSarima = emissionMain.copy()
dropColsArima = ['Coal, Including Coal Coke Net Imports, CO2 Emissions','Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions','Distillate Fuel Oil, Excluding Biodiesel, CO2 Emissions','Motor Gasoline, Excluding Ethanol, CO2 Emissions','Residual Fuel Oil CO2 Emissions','Petroleum, Excluding Biofuels, CO2 Emissions','Other Carbon Emission']
emissionSarima = emissionSarima.drop(dropColsArima, axis=1)
emissionSarima = emissionSarima.set_index('Month')
emissionSarima
```

```python
percent_training = 0.85
splitPoint = round(len(emissionSarima) * percent_training)
training, testing = emissionSarima[0:splitPoint], emissionSarima[splitPoint:]
```

```python
training = np.log(training)
training_diff = training.diff(periods=1).values[1:]

plt.plot(training_diff)
```

```python
#ACF and PACF plots
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(training_diff, nlags=40)
lag_pacf = pacf(training_diff, nlags=40, method='ols')
#ols is Ordinary Least Squared method
#Plot ACF:
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.stem(lag_acf)
# plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.stem(lag_pacf)
plt.axhline(y=0, linestyle='--', color='grey')
plt.axhline(y=-1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(training_diff)), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(training, order=(4,1,4), seasonal_order=(1,0,0,12), enforce_stationarity=False, enfoce_invertibility=False)
model_fit = model.fit(disp=False)
```

```python
K  = len(testing)
forecast = model_fit.forecast(K)
forecast = np.exp(forecast)

plt.figure(figsize=(20,10))
plt.plot(forecast,'r')
plt.plot(emissionSarima,'b')
plt.autoscale(enable=True, axis='x', tight=True)
plt.axvline(x=emissionSarima.index[splitPoint], color='black');
```

```python
predict = model_fit.predict(24,588)
exp_predict = np.exp(predict)
exp_predict

plt.figure(figsize=(20,10))
plt.plot(emissionSarima)
plt.plot(exp_predict)
```
